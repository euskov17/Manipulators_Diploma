```python
PI_CONSTANT = np.pi
DOUBLE_PI_CONSTANT = 2 * np.pi
```

```python
class Manipulator_2d_node():
    def __init__(self, angles, g = 0, h = 0, f = None, parent = None, path_from_parent=None):
        self.angles = angles
        self.g = g
        self.h = h
        if f is None:
            self.f = self.g + h
        else:
            self.f = f        
        self.parent = parent

    def get_angles(self):
        return self.angles
    
    def __eq__(self, other):
        return (self.angles == other.angles).all()
    
    def __hash__(self):
        return hash(tuple(self.angles))

    def __lt__(self, other): 
        return self.f < other.f
```

```python
class Manipulator_2d_supervisor():
    def __init__(self, num_joints, lengths, delta_numbers, angles_constraints, possible_moves=None, end_touch_radius=0.5, ground_level=0):
        self.num_joints = num_joints
        self.lengths = lengths
        self.radius = sum(self.lengths)
        self.delta_numbers = delta_numbers
        self.deltas = DOUBLE_PI_CONSTANT / delta_numbers
        self.angles_constraints = angles_constraints
        if possible_moves == 'combined':
            lst_combined_moves = np.array([list(i) for i in itertools.product([0, 1, -1], repeat=self.num_joints)])[1:, :]
            lst_solo_moves = np.array([[0] * (self.num_joints - i) + [1] * i for i in range(1, self.num_joints + 1)])
            self.move_costs = np.sum(np.abs(lst_combined_moves), axis=1)
            self.moves = lst_combined_moves[:, ::-1] @ lst_solo_moves
        elif possible_moves == 'solo':
            solo_positive_moves = [[0] * (self.num_joints - i) + [1] * i for i in range(1, self.num_joints + 1)]
            solo_negative_moves = [[0] * (self.num_joints - i) + [-1] * i for i in range(1, self.num_joints + 1)]
            self.moves = np.array(solo_positive_moves + solo_negative_moves)
            self.move_costs = np.stack([self.deltas, self.deltas], axis = 0).flatten()
        self.end_touch_radius = end_touch_radius
        self.ground_level = ground_level

    def get_move_costs(self):
        return self.move_costs
    
    def calculate_dots(self, angles):
        dots = np.zeros((self.num_joints + 1, 2))
        sin = np.sin(angles) * self.lengths
        cos = np.cos(angles) * self.lengths
        dots[1:, 1] = np.cumsum(sin)
        dots[1:, 0] = np.cumsum(cos)
        return dots

    def calculate_end(self, angles):
        x = np.sum(np.sin(angles) * self.lengths)
        y = np.sum(np.cos(angles) * self.lengths)
        return (x, y)
    
    def calculate_distance_between_states(self, angles_1, angles_2):
        diffs = (np.abs(angles_1 - angles_2)) % DOUBLE_PI_CONSTANT
        angle_diffs = np.minimum(diffs, (DOUBLE_PI_CONSTANT - diffs))
        summ = angle_diffs.sum()
        return summ
    
    def generate_random_state(self):
        random_delta_numbers = np.random.randint(0, self.delta_numbers)
        random_state = (random_delta_numbers * PI_CONSTANT / self.delta_numbers) % DOUBLE_PI_CONSTANT
        while not self.position_correctness(random_state):
            random_delta_numbers = np.random.randint(0, self.delta_numbers)
            random_state = (random_delta_numbers * PI_CONSTANT / self.delta_numbers) % DOUBLE_PI_CONSTANT
        return random_state

    def circle_segment_intersection(self, center_coord: np.array, radius: float, segment_begin: np.array, segment_end: np.array):
        '''
        Returns True if NO intersection, else False
        '''
        v = segment_end - segment_begin
        a = np.dot(v, v)
        b = 2 * np.dot(v, segment_begin - center_coord)
        c = np.dot(segment_begin, segment_begin) + np.dot(center_coord, center_coord) - 2 * np.dot(segment_begin, center_coord) - radius ** 2
        disc = b ** 2 - 4 * a * c
        if disc < 0:
            return True
        sqrt_disc = np.sqrt(disc)
        t1 = (-b + sqrt_disc) / (2 * a)
        t2 = (-b - sqrt_disc) / (2 * a)
        if not (0 <= t1 <= 1 or 0 <= t2 <= 1):
            return True
        return False
    
    def end_is_not_touching_manupulator(self, dots):
        '''
        Returns True if end of manipulator DOESNT touches anything, else False
        '''
        end = dots[-1]
        for segment_1_index in range(self.num_joints - 2):
            for segment_2_index in range(segment_1_index + 1, self.num_joints - 1):
                if self.circle_segment_intersection(end, self.end_touch_radius, dots[segment_1_index], dots[segment_2_index]) == False:
                    return False
        return True

    def dot_on_segment(self, x1, y1, x2, y2, x3, y3): 
        '''
        checks if (x2, y2) lies in segment ((x1, y1), (x3, y3))
        '''
        if ((x2 <= max(x1, x3)) and (x2 >= min(x1, x3)) and 
            (y2 <= max(y1, y3)) and (y2 >= min(y1, y3))): 
            return True
        return False

    def orientation(self, x1, y1, x2, y2, x3, y3): 
        '''
        returns orientation of 3 given dots
        '''
        val = (y2 - y1) * (x3 - x2) - (x2 - x1) * (y3 - y2)
        if val > 0: 
            return 'clockwize'
        elif val < 0: 
            return 'counterclockwise'
        else: 
            return 'collinear'

    def do_intersect(self, x1, y1, x2, y2, x3, y3, x4, y4): 
        o1 = self.orientation(x1, y1, x2, y2, x3, y3) 
        o2 = self.orientation(x1, y1, x2, y2, x4, y4) 
        o3 = self.orientation(x3, y3, x4, y4, x1, y1) 
        o4 = self.orientation(x3, y3, x4, y4, x2, y2) 
        if (o1 != o2) and (o3 != o4): 
            return True
        if (o1 == 0) and self.dot_on_segment(x1, y1, x3, y3, x2, y2): 
            return True
        if (o2 == 0) and self.dot_on_segment(x1, y1, x4, y4, x2, y2): 
            return True
        if (o3 == 0) and self.dot_on_segment(x3, y3, x1, y1, x4, y4): 
            return True
        if (o4 == 0) and self.dot_on_segment(x3, y3, x2, y2, x4, y4): 
            return True
        return False

    def position_intersection(self, dots):
        '''
        Returns True if NO self-intersection, else False
        '''
        for segment_1_index in range(self.num_joints - 1):
            for segment_2_index in range(segment_1_index + 1, self.num_joints):
                if segment_1_index + 1 == segment_2_index:
                    continue
                segment_1_dots = dots[segment_1_index, 0], dots[segment_1_index, 1], dots[segment_1_index + 1, 0], dots[segment_1_index + 1, 1]
                segment_2_dots = dots[segment_2_index, 0], dots[segment_2_index, 1], dots[segment_2_index + 1, 0], dots[segment_2_index + 1, 1]
                if self.do_intersect(*segment_1_dots, *segment_2_dots):
                    return False
        return True

    def adjacent_joints_angles(self, angles):
        '''
        Returns True if angles are correct, else False
        '''
        if self.angles_constraints[0][0] < angles[0] < self.angles_constraints[1][1]:
            return False
        for angle_index in range(self.num_joints - 1):
            if self.angles_constraints[angle_index + 1][0] < (angles[angle_index] - angles[angle_index + 1]) % DOUBLE_PI_CONSTANT < self.angles_constraints[angle_index + 1][1]:
                return False
        return True

    def position_correctness(self, angles):
        dots = self.calculate_dots(angles)
        return self.position_intersection(dots) and self.adjacent_joints_angles(angles) and (dots[1:, 1] > self.ground_level).all() and self.end_is_not_touching_manupulator(dots)

    def get_successors(self, angles):
        '''
        Returns massive, which elements are [<elements of moves massive>, <move_cost>, <x of end>, <y of end>]
        '''
        possible_neighbours = (self.moves * self.deltas + angles) % DOUBLE_PI_CONSTANT
        successors = []
        for successor_index in range(len(possible_neighbours)):
            successor_state = possible_neighbours[successor_index]
            if self.position_correctness(successor_state) == True:
                successors.append([successor_state, self.move_costs[successor_index], *self.calculate_end(successor_state)])
        return successors

    def are_states_directly_connected(self, begin_state, end_state):
        connect_path = []
        current_state = begin_state
        while not self.calculate_distance_between_states(current_state, end_state) < 0.01:
            print(f' current = {current_state}')
            best_neighbour = None
            for successor_state, move_cost, x, y in self.get_successors(current_state):
                diffs = (np.abs(successor_state - end_state)) % DOUBLE_PI_CONSTANT
                angle_diffs = np.minimum(diffs, (DOUBLE_PI_CONSTANT - diffs))

                print(f'successor = {successor_state}, dist = {self.calculate_distance_between_states(successor_state, end_state)}, diff = {angle_diffs}')
                if self.calculate_distance_between_states(successor_state, end_state) < self.calculate_distance_between_states(current_state, end_state):
                    neighbour = Manipulator_2d_node(successor_state, parent=current_state) 
                    best_neighbour = neighbour
                    connect_path.append(best_neighbour)
                    current_state = successor_state     
            diffs = (np.abs(successor_state - end_state)) % DOUBLE_PI_CONSTANT
            angle_diffs = np.minimum(diffs, (2 * DOUBLE_PI_CONSTANT - diffs))
            print(f'chosen = {best_neighbour.get_angles()}, dist = {self.calculate_distance_between_states(best_neighbour.get_angles(), end_state)}, diff = {angle_diffs}')
            if best_neighbour is None:
                return False, None
        return True, connect_path
    
    def visualize_state(self, angles):
        dots = self.calculate_dots(angles)
        plt.figure(figsize=(10,6))
        plt.axis([-self.radius, self.radius, 0, self.radius])
        plt.plot(dots[:, 0], dots[:, 1])
        plt.scatter(dots[:, 0], dots[:, 1])
```

```python
class SearchTreePQS():
    def __init__(self):
        self._open = []
        heapq.heapify(self._open)
        self._closed = set()
        self._enc_open_dublicates = 0

    def open_is_empty(self):
        return len(self._open) == 0

    def add_to_open(self, item):
        heapq.heappush(self._open, item)

    def get_best_node_from_open(self):
        current = heapq.heappop(self._open)
        while current in self._closed:
            current = heapq.heappop(self._open)
        return current

    def add_to_closed(self, item):
        self._closed.add(item)

    def was_expanded(self, item):
        return item in self._closed

    @property
    def OPEN(self):
        return self._open
    
    @property
    def CLOSED(self):
        return self._closed

    @property
    def number_of_open_dublicates(self):
        return self._enc_open_dublicates
    
```

```python
def make_path(goal):
    '''
    Creates a path by tracing parent pointers from the goal node to the start node
    It also returns path's length.
    '''

    length = goal.g
    current = goal
    path = []
    while current.parent:
        path.append(current)
        current = current.parent
    path.append(current)
    return path[::-1], length
```

```python
def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
```

```python
def astar(supervisor: Manipulator_2d_supervisor, manipulator: Manipulator_2d, goal_x, goal_y, map=None, heuristic_func=None, search_tree=None):
    ast = search_tree()
    steps = 0
    nodes_created = 0
    CLOSED = None
    
    start_position = manipulator.get_dots()[-1]
    start_angles = manipulator.get_angles()

    start = Manipulator_2d_node(start_angles, 0, 0, heuristic_func(start_position[0], start_position[1], goal_x, goal_y))
    ast.add_to_open(start)
        
    current_state = start
    while not heuristic_func(*supervisor.calculate_end(current_state.get_angles()), goal_x, goal_y) < 0.5:
        current_state = ast.get_best_node_from_open()
        for successor_state, move_cost, x, y in supervisor.get_successors(current_state.get_angles()):
            neighbour = Manipulator_2d_node(successor_state, current_state.g + 1 / 180 * np.pi * move_cost, heuristic_func(x, y,  goal_x, goal_y) / 2, parent=current_state)
            nodes_created += 1
            if not ast.was_expanded(neighbour):
                ast.add_to_open(neighbour) 
        ast.add_to_closed(current_state)
        steps += 1
        
    OPEN = ast.OPEN 
    CLOSED = ast.CLOSED
    return True, current_state, steps, nodes_created, OPEN, CLOSED

```

```python
def create_rrt(supervisor: Manipulator_2d_supervisor, start_state, near_radius, num_steps):
    added_nodes = [Manipulator_2d_node(start_state)]
    num_nodes = 1
    while num_nodes < num_steps:
        new_state = supervisor.generate_random_state()
        
        nearest_node = added_nodes[0]
        for selected_node in added_nodes:
            if sup.calculate_distance_between_states(selected_node.get_angles(), new_state) < sup.calculate_distance_between_states(nearest_node.get_angles(), new_state):
                nearest_node = selected_node
                
        if sup.calculate_distance_between_states(nearest_node.get_angles(), new_state) >= near_radius:
            continue
        
        connection, path = supervisor.are_states_directly_connected(nearest_node.get_angles(), new_state)
        print(f'new_state = {new_state / PI_CONSTANT}')
        print(f'nearest node = {selected_node.get_angles() / PI_CONSTANT}')
        print(f'diff = {(new_state - selected_node.get_angles()) / PI_CONSTANT / (3/180)}')
        print(f'connection = {connection}')
        if connection == False:
            continue
        else:
            num_nodes += 1
            print(num_nodes)
            added_nodes.append(Manipulator_2d_node(new_state, parent=nearest_node, path_from_parent=path))
    return added_nodes
```
